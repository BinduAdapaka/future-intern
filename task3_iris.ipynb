```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plot histogram for 'Age' column
data = pd.read_csv('Iris.csv')
```

```python

# Function to plot bar chart for a given column
def plot_bar_chart(column_name):

    avg_values = data.groupby('Species')[column_name].mean()

    # Plot the bar chart
    plt.figure(figsize=(8, 6))
    avg_values.plot(kind='bar', color='yellow')
    plt.title(f'Average {column_name} per Species')
    plt.ylabel(f'{column_name} (cm)')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()


plot_bar_chart('SepalWidthCm')
plot_bar_chart('PetalLengthCm')
plot_bar_chart('PetalWidthCm')
plot_bar_chart('SepalWidthCm')
```

```python

# Function to plot histogram for a given column
def plot_histogram(column_name, bins=10):
    plt.figure(figsize=(8, 6))
    plt.hist(data[column_name], bins=bins, color='lightgreen', edgecolor='black')
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(f'{column_name} (cm)')
    plt.ylabel('Frequency')
    plt.grid(axis='y')
    plt.show()


plot_histogram('PetalLengthCm', bins=10)
plot_histogram('PetalWidthCm', bins=10)
plot_histogram('SepalLengthCm', bins=10)
plot_histogram('SepalWidthCm', bins=10)

```
